```python
import pandas as pd
import requests
import re

from bs4 import BeautifulSoup
```

```python
url = 'https://www.billboard.com/charts/hot-100/'
page = requests.get(url)
page.status_code
```

```python
soup = BeautifulSoup(page.content, 'html.parser') 
```

```python
# Create an empty DataFrame to store the data
data = {'title': [], 'artist': []}
df = pd.DataFrame(data)
```

```python
items = soup.find_all('li', class_='o-chart-results-list__item')

# Loop through the found items
for item in items:
    # Extract the title and artist information if present
    title_elem = item.find('h3', class_='c-title')
    artist_elem = item.find('span', class_='c-label')
    
    # Check if both elements are found
    if title_elem and artist_elem:
        title = title_elem.text.strip()
        artist = artist_elem.text.strip()
        
        # Append data to the DataFrame
        df = df.append({'title': title, 'artist': artist}, ignore_index=True)

# Change the index to start from 1
df.index = df.index + 1

```
